```python
import json
import matplotlib.pyplot as plt
```

```python
def plot_alphas(json_file):
    # Load JSON data
    with open(json_file) as f:
        data = json.load(f)

    # Extract third elements
    alphas = [entry[2] for entry in data]

    # Create plot
    plt.figure(figsize=(7, 5))
    plt.plot(alphas, label="Alpha loss")
    plt.xlabel("Optimization steps (smoothed to 1000 steps)")
    plt.ylabel("Alpha loss")
    plt.title("Log-Alpha losses during SAC training.")
    plt.grid()
    plt.tight_layout()

    # Show plot
    plt.savefig("alpha_loss.png", dpi=300)
    plt.show()
```

```python
path = "sac_hockey_tensorboard.json"
plot_alphas(path)
```
