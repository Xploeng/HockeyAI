```python
import json
import matplotlib.pyplot as plt

from tueplots import bundles
from tueplots.constants.color import rgb
```

```python
plt.rcParams.update(
    bundles.icml2022(
        column="half",
        nrows=1,
        ncols=1,
        usetex=True,
    )
)
```

```python
def plot_alphas(json_file):
    # Load JSON data
    with open(json_file) as f:
        data = json.load(f)

    # Extract third elements
    alphas = [entry[2] for entry in data]

    # Create plot
    plt.figure(figsize=(5, 4))
    plt.plot(alphas, label="Alpha loss", color=rgb.tue_darkblue, linewidth=2)
    plt.xlabel("Optimization steps (smoothed to 1000 steps)", fontsize=10)
    plt.ylabel("Alpha loss", fontsize=10)
    # plt.title("Log-Alpha losses during SAC training.", fontsize=14)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid()

    # Show plot
    plt.savefig("alpha_loss.pdf", dpi=500)
    plt.show()
```

```python
path = "sac_hockey_tensorboard.json"
plot_alphas(path)
```

```python
def plot_rewards(json_file):
    # Load JSON data
    with open(json_file) as f:
        data = json.load(f)

    # Extract third elements
    alphas = [entry[2] for entry in data]

    # Create plot
    plt.figure(figsize=(5, 4))
    plt.title("Running average of rewards during SAC training.", fontsize=14)
    plt.plot(alphas, label="Rewards")

    plt.xlabel("Optimization steps (smoothed to 1000 steps)", fontsize=12)
    plt.ylabel("Rewards", fontsize=12)
    plt.ylim(-0.08, -0.02)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)

    plt.grid()

    # Show plot
    plt.savefig("rewards_sac.pdf", dpi=500)
    plt.show()
```

```python
path = "sac_hockey_rewards_tensorboard.json"
plot_rewards(path)
```
