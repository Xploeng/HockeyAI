```python
import numpy as np

import matplotlib.pyplot as plt

# Initialize lists to store the timings
sample_times = []
transpose_times = []
compute_q_times = []
compute_loss_times = []
update_times = []
total_times = []

# Read the timings from the file
with open('timings.txt', 'r') as file:
    for line in file:
        if 'Sample time' in line:
            sample_times.append(float(line.split(': ')[1]))
        elif 'Transpose time' in line:
            transpose_times.append(float(line.split(': ')[1]))
        elif 'Compute Q time' in line:
            compute_q_times.append(float(line.split(': ')[1]))
        elif 'Compute loss time' in line:
            compute_loss_times.append(float(line.split(': ')[1]))
        elif 'Update Target time' in line:
            update_times.append(float(line.split(': ')[1]))
        elif 'Update priorities time' in line:
            total_times.append(float(line.split(': ')[1]))

# Calculate the average of each list
average_sample_time = np.mean(sample_times)
average_transpose_time = np.mean(transpose_times)
average_compute_q_time = np.mean(compute_q_times)
average_compute_loss_time = np.mean(compute_loss_times)
average_update_time = np.mean(update_times)
average_prio_time = np.mean(total_times)

# Print the averages
print(f'Average Sample time: {average_sample_time}')
print(f'Average Transpose time: {average_transpose_time}')
print(f'Average Compute Q time: {average_compute_q_time}')
print(f'Average Compute Loss time: {average_compute_loss_time}')
print(f'Average Update Target time: {average_update_time}')
print(f'Average Update priorities time: {average_prio_time}')

# Plot the average timings
labels = ['Sample time', 'Transpose time', 'Compute Q time', 'Compute Loss time', 'Update Target time', 'Update priorities time']
averages = [average_sample_time, average_transpose_time, average_compute_q_time, average_compute_loss_time, average_update_time, average_prio_time]

plt.figure(figsize=(12, 8))
plt.bar(labels, averages)
plt.xlabel('Operation')
plt.ylabel('Average Time (seconds)')
plt.title('Average Timings from timings.txt')
plt.show()
```
